# Lendo uma imagem

```python
img = cv.imread("../imagens/messi5.jpg")

print(img.shape)
```

Outros parâmetros da função imread, além do nome do arquivo, é a flag que pode ser:
1. `cv2.IMREAD_GRAYSCALE` ou `0`: Carrega a imagem na escala de cinza
2. `cv2.IMREAD_COLOR`ou `1`: Carrega a imagem colorida. É o parâmetro padrão da função.
3. `cv2.IMREAD_UNCHANGED` ou `-1`: Carrega a imagem incluindo o canal alpha.

```python
#Para plotar imagens com matplotlib podemos fazer dessa forma, mas fique atento porque o matplotlib trabalha com colormaps.
fig = plt.figure(figsize=(5,5))
plt.imshow(img[:,:,::-1])
```

```python
fig = plt.figure(figsize=(5,5))
plt.imshow(img[::-1,:,::-1])
```

```python
fig = plt.figure(figsize=(5,5))
plt.imshow(img[:,::-1,::-1])
```

# Splitting and Merging (canais de cores)

```python
img = cv.imread("../imagens/pao_de_acucar.jpg")
```

```python
fig = plt.figure(figsize=(5,5))
plt.imshow(img[:,:,::-1])
```

```python
b, g, r = cv.split(img)

fig, ax = plt.subplots(1,4,figsize=(20,5))
ax[0].imshow(b, cmap="gray")
ax[1].imshow(g, cmap= "gray")
ax[2].imshow(r, cmap= "gray")

img_merged = cv.merge((b,g,r))

ax[3].imshow(img_merged[:, :, ::-1])
```

```python
ball = img[280:340, 330:390]
img[273:333, 100:160] = ball

cv.imwrite('messi_bola_ducplicada.png', img)

cv.imshow("Janela", img)
cv.waitKey(0)
cv.destroyAllWindows()
```
