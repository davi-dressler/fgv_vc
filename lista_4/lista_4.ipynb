```python
def transladar_imagem(img, centro):
    h, w = img.shape[:2]
    x, y = centro
    
    img_transladada = np.zeros_like(img)
    for i in range(h):
        for j in range(w):
            x_new = i + x
            y_new = j + y
            
            if(x_new < h and y_new < w):
                img_transladada[x_new, y_new] = img[i, j]
                
    return img_transladada
```

```python
palazzo_transladado = transladar_imagem(palazzo, (150,300))
```
